# Aula 01
## Primeiras visualizações

```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/monitoramento_tempo.csv')
```

```python
df.head().round(2)
```

```python
# cheacando se há alguma linha nula
df.info()
```

```python
df.data = pd.to_datetime(df.data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(df.data, df.temperatura)
```

```python
# Configurando melhor

# criando figura
fig = plt.figure(figsize=(15,8))

# criando eixo
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df.data, df.temperatura, color='teal')

# definindo títulos
eixo.set_title('Temperatura X Tempo', fontsize=22)
eixo.set_ylabel('Temperatura (K)', fontsize=18)
eixo.set_xlabel('Ano', fontsize=18)

# legenda
eixo.legend(['temperatura'], loc='lower right', fontsize=14)
```

# Criando limitação de data

```python
from datetime import datetime
```

```python
# criando figura
fig = plt.figure(figsize=(15,8))

# criando eixo
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df.data, df.temperatura, color='teal')

# limite de data
eixo.set_xlim(datetime(2014,1,1),datetime(2015,1,1))

# definindo títulos
eixo.set_title('Temperatura X Tempo', fontsize=22)
eixo.set_ylabel('Temperatura (K)', fontsize=18)
eixo.set_xlabel('Ano', fontsize=18)

# legenda
eixo.legend(['temperatura'], loc='lower right', fontsize=14)
```

# Somente maio de 2014

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.68,0.65,0.3,0.3])

eixo.grid(True)
eixo.plot(df.data, df.temperatura, color='g')
eixo.set_xlim(datetime(2014,5,1), datetime(2014,6,1))
eixo.set_ylim(270,320)
eixo.set_title('Temperatura Maio de 2014', fontsize=22, pad=20)
eixo.set_ylabel('Temperatura (K)', fontsize=18)
eixo.set_xlabel('Ano', fontsize=18)
eixo.legend(['temperatura'], loc='lower right', fontsize=14)

# cores diferentes no mesmo gráfico
azul_left = df.data < datetime(2014,5,1)  
azul_right = df.data > datetime(2014,6,1)

eixo2.grid(True)
eixo2.plot(df.data, df.temperatura, color='g')
eixo2.plot(df[azul_left]['data'], df[azul_left]['temperatura'], color='teal')
eixo2.plot(df[azul_right]['data'], df[azul_right]['temperatura'], color='teal')
eixo2.set_xlim(datetime(2014,1,1), datetime(2015,1,1))
eixo2.set_title('Temperatura em 2014', fontsize=14)
eixo2.set_ylabel('Temperatura (K)', fontsize=12)
eixo2.set_xlabel('Ano', fontsize=12)
eixo2.legend(['temperatura'], loc='best', fontsize=13)
```

# Aula 03
## Anotações e diversos gráficos

# Adicionando linhas limitadoras

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color='teal')
eixo.set_title('Temperatura X Tempo', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura (K)', fontsize=20)
eixo.legend(['Temperatura'], loc='best', fontsize=15)
eixo.grid(True)

# linhas min e max
eixo.axhline(max(df['temperatura']), color='k', linestyle='--')
eixo.axhline(min(df['temperatura']), color='k', linestyle='--')
```

# Adicionando Notações

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color='teal')
eixo.set_title('Temperatura X Tempo', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura (K)', fontsize=20)
eixo.legend(['Temperatura'], loc='best', fontsize=15)
eixo.grid(True)
eixo.axhline(max(df['temperatura']), color='k', linestyle='--')
eixo.axhline(min(df['temperatura']), color='k', linestyle='--')

# MÁXIMO

# local que a seta aponta
x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

# local do texto
x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura'] - 5)

# texto e seta
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, 
              xytext=(x2,y2), arrowprops=dict(facecolor='k'))

# MÍNIMO

# local que a seta aponta
x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])

# local do texto
x2 = df['data'][df['temperatura'].idxmin() - 7000]
y2 = min(df['temperatura'] + 5)

# texto e seta
eixo.annotate('Mínimo', xy=(x1,y1), fontsize=20, 
              xytext=(x2,y2), arrowprops=dict(facecolor='k'))
```

# Temperatura por dia da semana

```python
# group_by
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
# ordem dias
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]

temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))
cores = ['r', 'orange', 'yellow', 'g', 'b', 'indigo', 'purple']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor='k')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

# Gráfico de Pizza

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))
cores = ['r', 'orange', 'yellow', 'g', 'b', 'indigo', 'purple']

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
         autopct='%.1f%%')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

# Explodir

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

indice = range(len(temperatura_por_dia_da_semana))
cores = ['r', 'orange', 'yellow', 'g', 'b', 'indigo', 'purple']

eixo.pie(temperatura_por_dia_da_semana, 
         labels=temperatura_por_dia_da_semana.index,
         autopct='%.1f%%',  # porcentagem
         explode=explodir, shadow=True)  # explodir
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```
