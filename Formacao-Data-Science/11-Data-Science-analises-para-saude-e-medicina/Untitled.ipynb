```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df.head()
```

```python
# valor falso pois o erro Ã© muito grande
# nÃ£o vou considerar essa anÃ¡lise pois tem muito campo vazio (muita disparidade no padrÃ£o)
df['Country/Region'].value_counts()
```

```python
# Agrupando por paÃ­s
confirmed_by_country = df.groupby('Country/Region').sum(numeric_only=True)
confirmed_by_country.sample(10)
```

```python
confirmed_by_country[confirmed_by_country.index.str.contains('China')]
```

```python
import matplotlib.pyplot as plt
```

### plot de casos confirmados ao longo dos dias na China

```python
# Mostrando os dados de uma forma "enganativa"

plt.figure(figsize=(2,5))
plt.title('Casos confirmados na China', fontsize=16)

confirmed_by_country.loc['China'][2:43].plot()

plt.show()
```

```python
# DATA VISUALIZATION Ã© fundamental
# atenÃ§Ã£o aos 2 eixos

plt.figure(figsize=(10,5))
plt.title('Casos confirmados na China', fontsize=16)

confirmed_by_country.loc['China'][2:43].plot()

plt.show()
```

```python
# mais grÃ¡fico errado
plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:43].plot()
plt.title("Casos confirmados na 'China'")
plt.ylim(40000, 80000)  # somete a partir de 40.000 casos
plt.show()
```

```python
# plotando os dias finais

plt.figure(figsize=(10,5))
plt.title('Casos confirmados na China', fontsize=16)

confirmed_by_country.loc['China'][25:43].plot()

plt.show()
```

```python
# Novos casos
new_cases_china = confirmed_by_country.loc['China'][2:43].diff().dropna()
```

```python
plt.figure(figsize=(10,5))
plt.title('Novos casos na China', fontsize=16)

new_cases_china.plot()

plt.show()
```

```python
# last dates

plt.figure(figsize=(10,5))
plt.title('Confirmados por PaÃ­s (Ãºltimo dia)', fontsize=16)
last_date = confirmed_by_country.columns[-1] 
confirmed_by_country[last_date].plot(kind='bar')

plt.show()
```
