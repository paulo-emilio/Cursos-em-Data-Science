# Anlisando as notas do Movielens

```python
notas = pd.read_csv('../dados/ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby('movieId').mean().rating
nota_media_por_filme.head()
```

```python
ax = sns.histplot(nota_media_por_filme.values, kde=True)
ax.grid()
ax.set(xlabel='Nota Média', 
       ylabel='Frequência', 
       title='Média de votos em filmes no Movielens 100k')

ax
```

```python
# contagem de votos por filme
qtd_votos_por_filme = notas.groupby('movieId').count()
# somente >= 10
notas_mais_de_10_votos = qtd_votos_por_filme.query('userId >= 10').index
notas_mais_de_10_votos.values
```

```python
# nota média por filmes que possuem 10 ou mais votos
nota_media_pelo_menos_10_votos = nota_media_por_filme.loc[notas_mais_de_10_votos.values]
nota_media_pelo_menos_10_votos.head()
```

```python
ax = sns.histplot(nota_media_pelo_menos_10_votos.values, kde=True)
ax.grid()
ax.set(xlabel='Nota Média', 
       ylabel='Frequência', 
       title='Média de votos em filmes no Movielens 100k (10 ou mais votos)')
ax
```

```python
ax = sns.boxplot(x=nota_media_pelo_menos_10_votos.values)
ax.set(xlabel='Nota média do filme', title='Média de votos em filmes no Movielens 100k (10 ou mais votos)')
ax
```

## Cumulative

```python
ax = sns.histplot(nota_media_pelo_menos_10_votos.values, 
                  kde=True, cumulative=True, stat="percent")
ax.grid()
ax.set(xlabel='Nota Média', 
       ylabel='% acumulada de filmes', 
       title='Média de votos em filmes no Movielens 100k (10 ou mais votos)')
ax
```

# Analisando a distribuição dos dados de outros campos do TMDB

## Contagem de votos

```python
ax = sns.histplot(df_mais_de_10_votos.vote_count, kde=True)
ax.grid()
ax.set(xlabel='Número de votos', 
       ylabel='Frequência', 
       title='Número de votos em filmes no TMDB 5k (10 ou mais votos)')
ax
```

## Orçamento

```python
budget_maior_0 = df.query('budget > 0').budget

ax = sns.histplot(budget_maior_0 * 1e-6, kde=True)
ax.grid()
ax.set(xlabel='Orçamento em milhões', 
       ylabel='Frequência', 
       title='Orçamento em filmes no TMDB 5k')
ax
```

## Popularidade

```python
popularity = df.query('popularity > 0').popularity  

ax = sns.histplot(popularity, kde=True)
ax.grid()
ax.set(xlabel='Popularidade', 
       ylabel='Frequência', 
       title='Popularidade em filmes no TMDB 5k')
ax
```

## Tempo de duração

```python
runtime = df.query('runtime > 0').runtime.dropna()

ax = sns.histplot(runtime, kde=True)
ax.grid()
ax.set(xlabel='Tempo de duração', 
       ylabel='Frequência', 
       title='Tempo de duração em filmes no TMDB 5k')
ax
```

```python
# Cumulativa

runtime = df.query('runtime > 0').runtime.dropna()

ax = sns.histplot(runtime, kde=True, 
                  cumulative=True, stat='percent')
ax.grid()
ax.set(xlabel='Tempo de duração', 
       ylabel='% de filmes', 
       title='Tempo de duração em filmes no TMDB 5k')
ax
```
